```python
x = df.drop('Class', axis=1)
y = df['Class']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
Dtc = DecisionTreeClassifier()
Dtc.fit(x_train, y_train)
```

```python
y_pred = Dtc.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
cm = confusion_matrix(y_test,y_pred)
print(cm)
print("Accuracy=",accuracy_score(y_test,y_pred))
print("Precision=",precision_score(y_test,y_pred))
print("Recall=",recall_score(y_test,y_pred))
print("F1 Score=",f1_score(y_test,y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)
acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred)
rec = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("Confusion Matrix:\n", cm)

metrics = {'Accuracy': acc, 'Precision': prec, 'Recall': rec, 'F1 Score': f1}

plt.figure(figsize=(6,5))
plt.bar(metrics.keys(), metrics.values(), color=['blue','orange','green','red'])
plt.ylim(0,1)
plt.title("Model Performance Metrics")
plt.ylabel("Score")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

results = {
    "Logistic Regression": {"Accuracy": 0.9986, "Precision": 0.6111 , "F1 Score": 0.5851},
    "Decision Tree": {"Accuracy": 0.9989, "Precision": 0.6785, "F1 Score": 0.7238},
    "KNeighbors": {"Accuracy": 0.9983, "Precision": 1.0, "F1 Score": 0.0970},
    "SVM": {"Accuracy": 0.9984, "Precision": 0.6041, "F1 Score": 0.2959}
}

models = list(results.keys())
metrics = ["Accuracy", "Precision", "F1 Score"]

# Build score matrix
scores = np.array([[results[m][metric] for metric in metrics] for m in models])

# Positions
x = np.arange(len(models))
bar_width = 0.25

plt.figure(figsize=(10,6))

# Plot bars for each metric
for i, metric in enumerate(metrics):
    plt.bar(x + i*bar_width, scores[:, i], width=bar_width, label=metric)

# Formatting
plt.xticks(x + bar_width, models, rotation=20)
plt.ylabel("Score")
plt.title("Algorithm Comparison: Accuracy, Precision, F1 Score")
plt.ylim(0,1.05)
plt.legend()
plt.show()
```
